```python
import os
import json
import re
import csv
from collections import defaultdict

# Thay đường dẫn gốc chứa các folder object
root = r"E:\LET_ME_COOK\Captone\PixelNerf_finetuning\dataset_pottery"

# Nếu cần support thêm CAM khác, thêm vào dict này: 'CAMn': offset
# Ví dụ: CAM1 -> offset 0, CAM2 -> offset 90 (vì 90 frame mỗi camera)
cam_offsets = {
    "CAM1": 0,
    "CAM2": 90,
    # "CAM3": 180,  # nếu có
}

# Regex để tìm số frame cuối trong tên file (ví dụ: ...0001.png -> 1)
num_pattern = re.compile(r"(\d+)(?=\.[Pp][Nn][Gg]$)")
# Regex để detect CAM label (CAM1, CAM2, ...)
cam_pattern = re.compile(r"(CAM\d+)", re.IGNORECASE)

report = {
    "total_objects": 0,
    "processed": 0,
    "full_180_and_json": [],
    "missing_json": [],
    "no_images": [],
    "incomplete_images": [],
    "extra_images": [],
    "rename_failures": {},
    "details": {}
}

def safe_int(s):
    try:
        return int(s)
    except:
        return None

for obj in sorted(os.listdir(root)):
    obj_path = os.path.join(root, obj)
    if not os.path.isdir(obj_path):
        continue

    report["total_objects"] += 1
    details = {
        "has_json": False,
        "json_count": 0,
        "png_count": 0,
        "renamed_count": 0,
        "skipped_count": 0,
        "errors": []
    }

    json_path = os.path.join(obj_path, "transforms.json")
    if not os.path.exists(json_path):
        report["missing_json"].append(obj)
        details["errors"].append("missing transforms.json")
        report["details"][obj] = details
        continue

    # load transforms.json
    try:
        with open(json_path, "r", encoding="utf-8") as f:
            data = json.load(f)
    except Exception as e:
        details["errors"].append(f"json load error: {e}")
        report["rename_failures"][obj] = details["errors"]
        report["details"][obj] = details
        continue

    frames = data.get("frames", [])
    expected = [os.path.basename(frame.get("file_path", "").replace("./","")) for frame in frames if "file_path" in frame]
    expected_set = set(expected)
    details["has_json"] = True
    details["json_count"] = len(expected)

    pngs = [f for f in os.listdir(obj_path) if f.lower().endswith(".png")]
    details["png_count"] = len(pngs)

    if len(pngs) == 0:
        report["no_images"].append(obj)
        details["errors"].append("no png images found")
        report["details"][obj] = details
        continue

    # Build expected_map: index -> expected_name (index is numeric index in JSON after shifting)
    expected_map = {}
    for name in expected:
        m = num_pattern.search(name)
        if not m:
            base = os.path.splitext(name)[0]
            n = safe_int(base)
            if n is None:
                details["errors"].append(f"cannot parse number from expected name: {name}")
                continue
            idx = n
        else:
            idx = int(m.group(1))
        expected_map[idx] = name

    # Prepare found_map: for each png find its numeric part AND cam (if any)
    found_entries = []  # list of tuples (filename, numeric_part, cam_label_or_None)
    for p in pngs:
        m_num = num_pattern.search(p)
        num = int(m_num.group(1)) if m_num else None

        m_cam = cam_pattern.search(p)
        cam_label = m_cam.group(1).upper() if m_cam else None

        found_entries.append((p, num, cam_label))

    # Build mapping from computed target index -> source filenames (list)
    # If cam_label present and in cam_offsets, target_idx = num + cam_offsets[cam_label]
    found_map = {}
    for filename, num, cam_label in found_entries:
        if num is None:
            # map to None key for unmapped
            found_map.setdefault(None, []).append(filename)
            continue

        offset = 0
        if cam_label and cam_label.upper() in cam_offsets:
            offset = cam_offsets[cam_label.upper()]
        # compute target index (1-based)
        target_idx = num + offset
        found_map.setdefault(target_idx, []).append(filename)

    # Now attempt renaming using expected_map keys (expected indices)
    rename_errors = []
    renamed = 0
    skipped = 0
    used_targets = set()

    for idx, target_name in expected_map.items():
        candidates = found_map.get(idx, [])
        if not candidates:
            rename_errors.append(f"missing source for expected index {idx:04d} -> target {target_name}")
            continue

        # choose best candidate: prefer one already named correctly, else prefer one containing CAM label,
        # else take first
        src_name = None
        if len(candidates) == 1:
            src_name = candidates[0]
        else:
            chosen = None
            for c in candidates:
                if c == target_name:
                    chosen = c; break
            if chosen is None:
                # prefer candidate with CAM in name
                for c in candidates:
                    if cam_pattern.search(c):
                        chosen = c; break
            if chosen is None:
                chosen = candidates[0]
            src_name = chosen

        src_path = os.path.join(obj_path, src_name)
        dst_path = os.path.join(obj_path, target_name)

        # if already correct, skip
        if os.path.normcase(src_path) == os.path.normcase(dst_path):
            skipped += 1
            used_targets.add(target_name)
            continue

        # handle existing destination (conflict)
        if os.path.exists(dst_path):
            if os.path.normcase(dst_path) == os.path.normcase(src_path):
                skipped += 1
                used_targets.add(target_name)
                continue
            else:
                tmp_backup = os.path.join(obj_path, f".bak_{target_name}")
                try:
                    os.rename(dst_path, tmp_backup)
                    os.rename(src_path, dst_path)
                    try:
                        os.remove(tmp_backup)
                    except:
                        pass
                    renamed += 1
                    used_targets.add(target_name)
                except Exception as e:
                    rename_errors.append(f"conflict rename error for {src_name} -> {target_name}: {e}")
                    # try restore
                    try:
                        if os.path.exists(tmp_backup) and not os.path.exists(dst_path):
                            os.rename(tmp_backup, dst_path)
                    except:
                        pass
                continue
        else:
            try:
                os.rename(src_path, dst_path)
                renamed += 1
                used_targets.add(target_name)
            except Exception as e:
                rename_errors.append(f"rename error {src_name} -> {target_name}: {e}")
                continue

    # After renames, recompute
    final_pngs = [f for f in os.listdir(obj_path) if f.lower().endswith(".png")]
    details["renamed_count"] = renamed
    details["skipped_count"] = skipped

    exist_targets = [t for t in expected if os.path.exists(os.path.join(obj_path, t))]
    missing_targets = [t for t in expected if t not in exist_targets]
    if missing_targets:
        details["errors"].append(f"missing expected after rename: {len(missing_targets)} files")
        details["missing_expected_list"] = missing_targets[:20]

    extra = [p for p in final_pngs if p not in expected_set]
    if extra:
        details["extra_images_list"] = extra[:20]

    if details["json_count"] == 180 and len(exist_targets) == 180:
        report["full_180_and_json"].append(obj)
    else:
        if details["json_count"] == 0:
            report["missing_json"].append(obj)
        else:
            if len(exist_targets) < details["json_count"] or details["json_count"] != 180:
                report["incomplete_images"].append(obj)
            if len(extra) > 0:
                report["extra_images"].append(obj)

    if rename_errors:
        report["rename_failures"][obj] = rename_errors

    details["final_png_count"] = len(final_pngs)
    details["expected_present_count"] = len(exist_targets)
    report["details"][obj] = details
    report["processed"] += 1

# Save report files into root
out_json = os.path.join(root, "rename_report.json")
out_csv = os.path.join(root, "rename_report.csv")
with open(out_json, "w", encoding="utf-8") as f:
    json.dump(report, f, indent=2, ensure_ascii=False)

with open(out_csv, "w", newline='', encoding="utf-8") as csvf:
    writer = csv.writer(csvf)
    header = ["object","has_json","json_count","png_count","final_png_count","expected_present_count","renamed_count","skipped_count","errors"]
    writer.writerow(header)
    for obj, det in report["details"].items():
        writer.writerow([
            obj,
            det.get("has_json", False),
            det.get("json_count", 0),
            det.get("png_count", 0),
            det.get("final_png_count", 0),
            det.get("expected_present_count", 0),
            det.get("renamed_count", 0),
            det.get("skipped_count", 0),
            "; ".join(det.get("errors", [])) if det.get("errors") else ""
        ])

# Print summary
print("=== RENAME BY JSON SUMMARY (with CAM offsets) ===")
print(f"Total object folders found: {report['total_objects']}")
print(f"Processed folders: {report['processed']}")
print(f"Objects with full 180 images & valid json: {len(report['full_180_and_json'])}")
print(f"Objects missing json: {len(report['missing_json'])} -> {report['missing_json'][:30]}")
print(f"Objects with incomplete images / mismatch: {len(report['incomplete_images'])} -> {report['incomplete_images'][:30]}")
print(f"Objects with extra images: {len(report['extra_images'])} -> {report['extra_images'][:30]}")
print(f"Objects with rename failures: {len(report['rename_failures'])}")
if report['rename_failures']:
    for k, v in list(report['rename_failures'].items())[:20]:
        print(f" - {k}: {len(v)} error(s). Example: {v[:2]}")
print()
print(f"Saved detailed report to: {out_json}")
print(f"Saved summary CSV to: {out_csv}")
```

```python
import os
from PIL import Image

# ===========================
# CONFIG
# ===========================
root = r"E:\LET_ME_COOK\Captone\PixelNerf_finetuning\dataset_pottery_ver3"   # thư mục chứa các object
target_size = (128, 128)   # resize về 128x128
overwrite = True           # True = ghi đè, False = lưu file mới (xxx_resized.png)

# ===========================
# PROCESS
# ===========================
count_obj = 0
count_img = 0

for obj in sorted(os.listdir(root)):
    obj_path = os.path.join(root, obj)
    if not os.path.isdir(obj_path):
        continue
    
    count_obj += 1
    print(f"[PROCESS] {obj}")

    for file in os.listdir(obj_path):
        if file.lower().endswith(".png"):
            img_path = os.path.join(obj_path, file)
            count_img += 1

            # Load ảnh
            try:
                img = Image.open(img_path).convert("RGBA")
            except:
                print(f"   [ERR] Cannot open: {file}")
                continue

            # Resize
            img_resized = img.resize(target_size, Image.LANCZOS)

            # Save
            if overwrite:
                save_path = img_path
            else:
                name, ext = os.path.splitext(img_path)
                save_path = f"{name}_resized{ext}"

            img_resized.save(save_path)

    print(f"   -> Done {obj}")

print("\n=== DONE ===")
print(f"Total objects processed: {count_obj}")
print(f"Total images resized: {count_img}")
```

```python
import shutil
from pathlib import Path

# ============ CONFIG ============
SRC_ROOT = Path(r"E:\LET_ME_COOK\Captone\PixelNerf_finetuning\dataset_pottery_128")
DST_ROOT = Path(r"E:\LET_ME_COOK\Captone\PixelNerf_finetuning\dataset_pottery_trans_128")

TRANS_FILENAME = "transforms.json"   # exact filename to copy
IMG_EXTS = {".png", ".jpg", ".jpeg"}
# ================================

def count_images_in_obj(obj_dir: Path) -> int:
    """
    Count image files in obj_dir and obj_dir/images (non-recursive).
    Returns an integer.
    """
    if not obj_dir.exists() or not obj_dir.is_dir():
        return 0
    total = 0
    # count files right inside obj_dir
    for p in obj_dir.iterdir():
        if p.is_file() and p.suffix.lower() in IMG_EXTS:
            total += 1
    # also count inside images/ if present
    images_dir = obj_dir / "images"
    if images_dir.exists() and images_dir.is_dir():
        for p in images_dir.iterdir():
            if p.is_file() and p.suffix.lower() in IMG_EXTS:
                total += 1
    return total

def main():
    if not SRC_ROOT.exists():
        print(f"Source root does not exist: {SRC_ROOT}")
        return
    if not DST_ROOT.exists():
        print(f"Destination root does not exist: {DST_ROOT}")
        return

    src_dirs = [p for p in sorted(SRC_ROOT.iterdir()) if p.is_dir()]

    copied = []
    skipped = []
    missing_dst = []
    missing_transform = []

    for src_dir in src_dirs:
        src_tf = src_dir / TRANS_FILENAME
        if not src_tf.exists():
            missing_transform.append(src_dir.name)
            print(f"[SKIP] {src_dir.name}: no {TRANS_FILENAME} in source.")
            continue

        dst_dir = DST_ROOT / src_dir.name
        if not dst_dir.exists() or not dst_dir.is_dir():
            missing_dst.append(src_dir.name)
            print(f"[SKIP] {src_dir.name}: destination folder not found -> {dst_dir}")
            continue

        # count images in both source and destination (including images/ subfolder)
        n_src = count_images_in_obj(src_dir)
        n_dst = count_images_in_obj(dst_dir)

        if n_src == n_dst:
            try:
                dst_tf = dst_dir / TRANS_FILENAME
                shutil.copy2(src_tf, dst_tf)
                copied.append(src_dir.name)
                print(f"[COPIED] {src_dir.name}: {TRANS_FILENAME} copied ({n_src} images each).")
            except Exception as e:
                skipped.append((src_dir.name, f"copy_error: {e}"))
                print(f"[ERROR] {src_dir.name}: failed to copy {TRANS_FILENAME}: {e}")
        else:
            skipped.append((src_dir.name, f"image_count_mismatch src={n_src} dst={n_dst}"))
            print(f"[SKIP] {src_dir.name}: image count mismatch (src={n_src}, dst={n_dst}). Not copied.")

    # Summary
    print("\n=== SUMMARY ===")
    print(f"Total source subfolders checked: {len(src_dirs)}")
    print(f"Copied: {len(copied)}")
    if copied:
        print(" -", ", ".join(copied))
    print(f"Skipped due to mismatch or error: {len(skipped)}")
    if skipped:
        for name, reason in skipped:
            print(f" - {name}: {reason}")
    print(f"Missing transform.json in source: {len(missing_transform)}")
    if missing_transform:
        print(" -", ", ".join(missing_transform))
    print(f"Missing destination folders: {len(missing_dst)}")
    if missing_dst:
        print(" -", ", ".join(missing_dst))

if __name__ == "__main__":
    main()
```

```python
import json
import math
from pathlib import Path
from PIL import Image
import shutil

ROOT = Path(r"E:\LET_ME_COOK\Captone\PixelNerf_finetuning\dataset_pottery")
IMG_EXTS = {".png", ".jpg", ".jpeg", ".bmp", ".webp"}

def normalize_file_path(rel: str) -> str:
    if rel is None:
        return ""
    rel = rel.replace("\\", "/")
    rel = rel.lstrip("./")
    if rel.startswith("images/"):
        rel = rel[len("images/"):]
    return rel.split("/")[-1]

def is_image_path(p: Path):
    return p.is_file() and p.suffix.lower() in IMG_EXTS

def read_image_size(obj_dir: Path, filename: str):
    """Trả về (w,h) nếu ảnh tồn tại trong obj_dir hoặc trong obj_dir/images"""
    if not filename:
        return None
    possible = [obj_dir / filename, obj_dir / "images" / filename]
    for p in possible:
        if p.exists() and p.is_file() and p.suffix.lower() in IMG_EXTS:
            try:
                with Image.open(p) as im:
                    return im.width, im.height
            except Exception:
                return None
    return None

def compute_camera_angle_x(w, fl_x):
    try:
        w = float(w)
        fl_x = float(fl_x)
        if fl_x == 0:
            return None
        return 2.0 * math.atan(w / (2.0 * fl_x))
    except Exception:
        return None

def process_folder(obj_dir: Path):
    tf_path = obj_dir / "transforms.json"
    if not tf_path.exists():
        print(f"  - [{obj_dir.name}] không có transforms.json, bỏ qua.")
        return

    try:
        with tf_path.open("r", encoding="utf-8") as f:
            tf = json.load(f)
    except Exception as e:
        print(f"  - [{obj_dir.name}] lỗi đọc transforms.json: {e}, bỏ qua.")
        return

    # sao lưu nếu chưa có backup
    bak = obj_dir / "transforms.json.bak"
    if not bak.exists():
        try:
            shutil.copy2(tf_path, bak)
        except Exception:
            pass

    # Lấy camera_angle_x nếu có
    cam_ax = tf.get("camera_angle_x", None)

    # Nếu không có camera_angle_x, cố tính từ fl_x và w (có thể ở root hoặc trong frame)
    if cam_ax is None:
        fl_x = tf.get("fl_x", tf.get("flx", None))
        w_root = tf.get("w", None)

        if fl_x is not None and w_root is not None:
            cam_ax = compute_camera_angle_x(w_root, fl_x)
        else:
            # try lấy từ frame đầu
            frames_orig = tf.get("frames", [])
            if frames_orig:
                fr0 = frames_orig[0]
                fl_x_fr = fr0.get("fl_x", fr0.get("flx", None))
                w_fr = fr0.get("w", None)
                if fl_x_fr is not None and w_fr is not None:
                    cam_ax = compute_camera_angle_x(w_fr, fl_x_fr)
                elif fl_x_fr is not None and w_root is not None:
                    cam_ax = compute_camera_angle_x(w_root, fl_x_fr)

    # Nếu vẫn None thì giữ None (khi ghi ra sẽ là null trong JSON)
    if cam_ax is not None:
        # ensure it's a float (json friendly)
        try:
            cam_ax = float(cam_ax)
        except Exception:
            cam_ax = None

    # Xử lý frames: chuẩn hoá file_path -> ./0001.png, lấy w,h và transform_matrix
    frames_new = []
    frames_orig = tf.get("frames", [])
    if not frames_orig:
        print(f"  - [{obj_dir.name}] transforms.json không có frames, bỏ qua.")
        return

    # Sắp theo thứ tự gốc để gán số thứ tự 0001...
    for idx, fr in enumerate(frames_orig, start=1):
        fname_raw = fr.get("file_path", "") or fr.get("file", "")
        fname = normalize_file_path(fname_raw)

        # tên mới theo yêu cầu
        ext = Path(fname).suffix.lower() if fname else ".png"
        new_fname = f"{idx:04d}{ext}"
        file_path_field = f"./{new_fname}"

        # lấy w/h ưu tiên: fr["w"], hoặc tf["w"], hoặc đọc ảnh thực tế
        w = fr.get("w", None)
        h = fr.get("h", None)
        if (w is None or h is None):
            # thử từ root tf
            if w is None:
                w = tf.get("w", w)
            if h is None:
                h = tf.get("h", h)

        if (w is None or h is None):
            # cố đọc ảnh gốc (vị trí fname trong obj_dir hoặc obj_dir/images)
            size = read_image_size(obj_dir, fname)
            if size:
                w, h = size

        # convert to int if possible
        try:
            w = int(w) if w is not None else None
        except Exception:
            w = None
        try:
            h = int(h) if h is not None else None
        except Exception:
            h = None

        # transform_matrix: nhiều dataset có key "transform_matrix" hoặc "transform"
        tm = fr.get("transform_matrix", fr.get("transform", None))
        # nếu tm là list phẳng hoặc dạng khác thì cố chuẩn hoá sang 4x4
        if tm is None:
            # một số transforms.json lưu pose dưới key 'matrix' hoặc 'transform_matrix'
            tm = fr.get("matrix", None)
        # nếu vẫn None, cố lấy toàn bộ frame dict làm transform (fallback)
        if tm is None:
            print(f"    ! [{obj_dir.name}] frame {idx} không có transform_matrix, sẽ để null.")
            tm = None

        frames_new.append({
            "file_path": file_path_field,
            "w": w,
            "h": h,
            "transform_matrix": tm
        })

    # Tạo dict mới
    out = {
        "camera_angle_x": cam_ax,
        "frames": frames_new
    }

    # Ghi file (ghi đè transforms.json)
    try:
        with tf_path.open("w", encoding="utf-8") as f:
            json.dump(out, f, ensure_ascii=False, indent=2)
        print(f"  + [{obj_dir.name}] đã chuẩn hoá và ghi transforms.json (backup tại transforms.json.bak).")
    except Exception as e:
        print(f"  - [{obj_dir.name}] Lỗi ghi transforms.json mới: {e}")

def main():
    if not ROOT.exists():
        print("ROOT không tồn tại:", ROOT)
        return

    folders = [p for p in sorted(ROOT.iterdir()) if p.is_dir()]
    if not folders:
        print("Không tìm thấy folder con trong ROOT.")
        return

    print(f"Tìm thấy {len(folders)} folder. Bắt đầu xử lý...\n")
    for d in folders:
        print(f"Xử lý folder: {d.name}")
        process_folder(d)
    print("\nHoàn tất.")

if __name__ == "__main__":
    main()
```

```python
import os
import json
import shutil
from pathlib import Path
import math
from PIL import Image

# ================ CẤU HÌNH ================
SRC_ROOT = Path(r"E:\LET_ME_COOK\Captone\PixelNerf_finetuning\dataset_pottery")
DST_ROOT = Path(r"E:\LET_ME_COOK\Captone\PixelNerf_finetuning\dataset_pottery_128")  # có thể cùng SRC_ROOT để chỉnh in-place
RESIZE_TO = (128, 128)
IMG_EXTS = {".png", ".jpg", ".jpeg", ".bmp", ".webp"}
# ===========================================

def normalize_file_path(rel: str) -> str:
    """Chuẩn hoá file_path trong transforms.json -> chỉ còn tên file."""
    if rel is None:
        return ""
    rel = rel.replace("\\", "/")
    rel = rel.lstrip("./")
    if rel.startswith("images/"):
        rel = rel[len("images/"):]
    return os.path.basename(rel)

def is_image_file(p: Path):
    return p.is_file() and p.suffix.lower() in IMG_EXTS

def read_image_size(obj_dir: Path, filename: str):
    """Trả về (w,h) nếu ảnh tồn tại trong obj_dir hoặc obj_dir/images"""
    if not filename:
        return None
    candidates = [obj_dir / filename, obj_dir / "images" / filename]
    for p in candidates:
        if p.exists() and p.is_file() and p.suffix.lower() in IMG_EXTS:
            try:
                with Image.open(p) as im:
                    return im.width, im.height
            except Exception:
                return None
    return None

def compute_scale_ratio(orig_w, target_w):
    try:
        orig_w = float(orig_w)
        if orig_w <= 0:
            return None
        return float(target_w) / orig_w
    except Exception:
        return None

def process_object(obj_dir: Path, dst_root: Path):
    """
    Xử lý 1 object:
      - đọc transforms.json
      - backup transforms.json -> transforms.json.bak (nếu chưa có)
      - load frames (theo thứ tự trong file)
      - resize mỗi ảnh (mở từ vị trí gốc) -> lưu vào dst_obj_dir/0001.ext, 0002.ext ...
      - cập nhật transforms.json: file_path, w, h, scale intrinsics nếu có
    """
    tf_path = obj_dir / "transforms.json"
    if not tf_path.exists():
        print(f"  ! [{obj_dir}] không có transforms.json, bỏ qua.")
        return False

    try:
        with tf_path.open("r", encoding="utf-8") as f:
            tf = json.load(f)
    except Exception as e:
        print(f"  ! [{obj_dir}] lỗi đọc transforms.json: {e}, bỏ qua.")
        return False

    # Backup nếu chưa có
    bak = obj_dir / "transforms.json.bak"
    if not bak.exists():
        try:
            shutil.copy2(tf_path, bak)
        except Exception:
            pass

    frames_orig = tf.get("frames", [])
    if not frames_orig:
        print(f"  ! [{obj_dir}] transforms.json không có frames, bỏ qua.")
        return False

    # map ảnh hiện có: tìm trong obj_dir hoặc obj_dir/images
    img_map = {}
    search_dirs = [obj_dir, obj_dir / "images"]
    for sd in search_dirs:
        if sd.exists() and sd.is_dir():
            for p in sd.iterdir():
                if p.is_file() and p.suffix.lower() in IMG_EXTS:
                    img_map[p.name] = p

    # Lấy orig_w để scale intrinsics nếu có
    orig_w = tf.get("w", None)
    if orig_w is None and len(frames_orig) > 0:
        orig_w = frames_orig[0].get("w", None)

    # Tạo dst object dir (giữ cấu trúc tương tự trong DST_ROOT)
    # Nếu DST_ROOT == SRC_ROOT thì đây chính là obj_dir (in-place)
    # Ta vẫn dùng dst_obj_dir để viết file
    relative_obj = obj_dir.relative_to(SRC_ROOT)
    dst_obj_dir = dst_root / relative_obj
    dst_obj_dir.mkdir(parents=True, exist_ok=True)

    new_frames = []
    # Duyệt theo thứ tự frames trong transforms.json, gán tên mới tuần tự
    for idx, fr in enumerate(frames_orig, start=1):
        fname_raw = fr.get("file_path", "") or fr.get("file", "")
        fname = normalize_file_path(fname_raw)

        src_img = img_map.get(fname)
        if src_img is None:
            # Nếu không tìm thấy theo tên chuẩn hoá, cố match bất kỳ file có tên chứa phần basename
            # (không bắt buộc nhưng hữu ích khi transforms.json có đường dẫn khác)
            for k, p in img_map.items():
                if k.lower().startswith(fname.lower().split('.')[0]) or fname.lower().startswith(k.lower().split('.')[0]):
                    src_img = p
                    break

        if src_img is None:
            print(f"    ! [{obj_dir.name}] frame #{idx}: không tìm thấy ảnh gốc '{fname}', sẽ bỏ frame này.")
            # Bỏ frame nếu không có file ảnh (không thêm vào new_frames)
            continue

        ext = src_img.suffix.lower()
        new_name = f"{idx:04d}{ext}"
        dst_img_path = dst_obj_dir / new_name

        # Mở và resize rồi lưu
        try:
            with Image.open(src_img) as im:
                im = im.convert("RGB")
                im_resized = im.resize(RESIZE_TO, Image.BILINEAR)
                im_resized.save(dst_img_path)
        except Exception as e:
            print(f"    ! [{obj_dir.name}] lỗi xử lý ảnh '{src_img}': {e}, bỏ frame.")
            continue

        # Lấy w,h: ưu tiên fr, sau đó tf root, sau đó đọc ảnh gốc (hoặc đọc kích thước đã resize)
        w = fr.get("w", None)
        h = fr.get("h", None)
        if w is None:
            w = tf.get("w", None)
        if h is None:
            h = tf.get("h", None)
        if w is None or h is None:
            # thử đọc kích thước ảnh gốc
            size = None
            try:
                with Image.open(src_img) as im:
                    size = (im.width, im.height)
            except Exception:
                size = None
            if size:
                w, h = size

        # Sau khi resize, override w/h bằng RESIZE_TO (đảm bảo nhất quán)
        w_resized, h_resized = RESIZE_TO
        w = int(w_resized)
        h = int(h_resized)

        # Clone frame dict và cập nhật
        fr_new = dict(fr)
        fr_new["file_path"] = f"./{new_name}"
        fr_new["w"] = w
        fr_new["h"] = h

        new_frames.append(fr_new)

    # Nếu không có frame nào hợp lệ sau xử lý thì bỏ
    if not new_frames:
        print(f"  ! [{obj_dir.name}] không còn frame hợp lệ sau xử lý, bỏ object.")
        return False

    # Tạo tf_new dựa trên tf gốc: giữ nguyên các khóa khác nhưng cập nhật frames, w, h
    tf_new = dict(tf)
    tf_new["frames"] = new_frames
    tf_new["w"] = RESIZE_TO[0]
    tf_new["h"] = RESIZE_TO[1]

    # Scale intrinsics nếu orig_w có hiệu lực
    if orig_w is not None:
        try:
            orig_w_val = float(orig_w)
            if orig_w_val > 0:
                scale_ratio = float(RESIZE_TO[0]) / orig_w_val
                for key in ("fl_x", "fl_y", "cx", "cy"):
                    if key in tf_new:
                        try:
                            tf_new[key] = float(tf_new[key]) * scale_ratio
                        except Exception:
                            # nếu parse float thất bại, bỏ qua
                            pass
        except Exception:
            pass

    # Ghi transforms.json mới (ghi đè file trong dst_obj_dir)
    dst_tf_path = dst_obj_dir / "transforms.json"
    try:
        with dst_tf_path.open("w", encoding="utf-8") as f:
            json.dump(tf_new, f, ensure_ascii=False, indent=2)
    except Exception as e:
        print(f"  ! [{obj_dir.name}] lỗi ghi transforms.json mới: {e}")
        return False

    print(f"  + [{obj_dir.name}] xử lý xong. Viết ảnh resized vào: {dst_obj_dir}")
    return True

def find_all_object_dirs(src_root: Path):
    """
    Tìm tất cả thư mục chứa transforms.json trong cây SRC_ROOT.
    Trả về list các Path (thư mục chứa transforms.json).
    """
    obj_dirs = []
    for tf in src_root.rglob("transforms.json"):
        parent = tf.parent
        obj_dirs.append(parent)
    # loại trùng
    obj_dirs = sorted(set(obj_dirs))
    return obj_dirs

def main():
    print("=== Chạy resize và chuẩn hóa transforms.json (không chọn N_FRAMES, không split) ===")
    print(f"SRC_ROOT = {SRC_ROOT}")
    print(f"DST_ROOT = {DST_ROOT}")
    print(f"Resize -> {RESIZE_TO}")
    print()

    if not SRC_ROOT.exists():
        print("SRC_ROOT không tồn tại:", SRC_ROOT)
        return

    obj_dirs = find_all_object_dirs(SRC_ROOT)
    if not obj_dirs:
        print("Không tìm thấy thư mục nào chứa transforms.json trong SRC_ROOT.")
        return

    print(f"Tìm thấy {len(obj_dirs)} object (thư mục chứa transforms.json). Bắt đầu xử lý...\n")
    n_ok = 0
    n_fail = 0
    for od in obj_dirs:
        print(f"Xử lý object: {od.relative_to(SRC_ROOT)}")
        ok = process_object(od, DST_ROOT)
        if ok:
            n_ok += 1
        else:
            n_fail += 1
        print()

    print("=== Hoàn tất ===")
    print(f"Thành công: {n_ok}, Thất bại: {n_fail}")

if __name__ == "__main__":
    main()
```
